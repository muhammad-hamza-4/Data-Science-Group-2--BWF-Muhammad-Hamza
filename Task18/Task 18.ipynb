# Introduction to Seaborn 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('Studentperformance.csv')

# first 30 student performance because data was it make it messy
subset = data.head(30)

sns.lineplot(x='StudyTimeWeekly', y='GPA', data=subset)
plt.title('STudent Performance')
plt.xlabel('Time')
plt.ylabel('GPA')
plt.show()


```

# Other Examples

# Scatter Plot

```python


data = sns.load_dataset('iris')


sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=data)

plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

# Line Plot

```python
# Load dataset
data = sns.load_dataset('flights')

sns.lineplot(x='year', y='passengers', hue='month', data=data)

plt.title('Number of Passengers Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.show()
```

# Bar Plot

```python

data = sns.load_dataset('titanic')

sns.barplot(x='class', y='survived', hue='sex', data=data)

plt.title('Survival Rate by Class and Sex')
plt.xlabel('Class')
plt.ylabel('Survival Rate')
plt.show()
```

# Count Plot

```python

da = sns.load_dataset('tips')

sns.countplot(x='day', hue='smoker', data=da)

plt.title('Count of Tips by Day and Smoker Status')
plt.xlabel('Day')
plt.ylabel('Count')
plt.show()
```

# Box Plot

```python

data = sns.load_dataset('diamonds')


sns.boxplot(x='cut', y='price', data=data)

plt.title('Box Plot of Diamond Prices by Cut')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.show()
```
