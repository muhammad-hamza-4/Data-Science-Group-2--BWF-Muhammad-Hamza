```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

np.random.seed(0)

#Ramdom DAta
data = pd.DataFrame({
    'x': np.random.normal(size=100),
    'y': np.random.normal(size=100)
})

```

# Linear Regression Plot 

```python

plt.figure(figsize=(8, 6))
sns.regplot(x='x', y='y', data=data)
plt.title('Linear Regression Plot')
plt.show()
```

# Multiple Linear Regression Plot 

```python

sns.lmplot(x='x', y='y', data=data)
plt.title('Multiple Linear Regression Plot')
plt.show()
```

# Residual Plot

```python

sns.residplot(x='x', y='y', data=data)
plt.title('Residual Plot')
plt.show()
```

# Histogram 

```python

plt.figure(figsize=(8, 6))
sns.histplot(data['x'], kde=False, bins=10)
plt.title('Histogram')
plt.show()
```

# Kernel Density Estimate Plot 

```python

plt.figure(figsize=(8, 6))
sns.kdeplot(data['x'], fill=True)
plt.title('KDE Plot')
plt.show()
```

# Distribution Plot 

```python

plt.figure(figsize=(8, 6))
sns.distplot(data['x'], hist=True, kde=True, bins=10)
plt.title('Distribution Plot')
plt.show()
```

# Violin Plot

```python

plt.figure(figsize=(8, 6))
sns.violinplot(y='y', data=data)
plt.title('Violin Plot')
plt.show()
```

# Heatmap 

```python

correlation_matrix = data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```

# Joint Plot

```python

sns.jointplot(x='x', y='y', data=data, kind='scatter')
plt.title('Joint Plot')
plt.show()
```

# Facet Grid 

```python

g = sns.FacetGrid(data, col='x')
g.map(sns.histplot, 'y')
plt.suptitle('Facet Grid')
plt.show()
```
