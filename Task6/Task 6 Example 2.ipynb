```python
import os

class Hotel:
    def __init__(self, name):
        self.name = name
        self.guests = {}
        self.file_name = 'guests.txt'
        self.load_guests()

    def load_guests(self):
        if os.path.exists(self.file_name):
            try:
                with open(self.file_name, 'r') as file:
                    for line in file:
                        guest_id, guest_name = line.strip().split(',')
                        self.guests[int(guest_id)] = guest_name
            except Exception as e:
                print(f"Error loading guests from file: {e}")

    def save_guests(self):
        try:
            with open(self.file_name, 'w') as file:
                for guest_id, guest_name in self.guests.items():
                    file.write(f"{guest_id},{guest_name}\n")
        except Exception as e:
            print(f"Error saving guests to file: {e}")

    def add_guest(self, guest_id, guest_name):
        try:
            if guest_id in self.guests:
                print(f"Guest with ID {guest_id} is already checked in.")
            else:
                self.guests[guest_id] = guest_name
                self.save_guests()
                print(f"Guest {guest_name} checked in with ID {guest_id}.")
        except Exception as e:
            print(f"Error adding guest: {e}")

    def checkout_guest(self, guest_id):
        try:
            if guest_id in self.guests:
                guest_name = self.guests.pop(guest_id)
                self.save_guests()
                print(f"Guest {guest_name} with ID {guest_id} has checked out.")
            else:
                print(f"No guest found with ID {guest_id}.")
        except Exception as e:
            print(f"Error checking out guest: {e}")

    def view_guests(self):
        try:
            if not self.guests:
                print("No guests currently checked in.")
            else:
                print("Current guests:")
                for guest_id, guest_name in self.guests.items():
                    print(f"ID: {guest_id}, Name: {guest_name}")
        except Exception as e:
            print(f"Error viewing guests: {e}")

def main():
    hotel = Hotel("Sunshine Hotel")
    
    while True:
        print("\nHotel Management System")
        print("1. Add Guest")
        print("2. Check Out Guest")
        print("3. View Guests")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            try:
                guest_id = int(input("Enter guest ID: "))
                guest_name = input("Enter guest name: ")
                hotel.add_guest(guest_id, guest_name)
            except ValueError:
                print("Invalid input. Please enter a valid guest ID.")

        elif choice == '2':
            try:
                guest_id = int(input("Enter guest ID to check out: "))
                hotel.checkout_guest(guest_id)
            except ValueError:
                print("Invalid input. Please enter a valid guest ID.")

        elif choice == '3':
            hotel.view_guests()

        elif choice == '4':
            print("Exiting the system. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number between 1 and 4.")

if __name__ == "__main__":
    main()
```
