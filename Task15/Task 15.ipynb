# Data Wrangling: Join, Combine, and Reshape.
New Question is seperated by Line Space

```python
import pandas as pd

df1 = pd.DataFrame({'key': ['F', 'B', 'C'], 'value1': [1, 2, 3]})
df2 = pd.DataFrame({'key': ['F', 'B', 'D'], 'value2': [4, 5, 6]})


mergedsinglekey = pd.merge(df1, df2, on='key')
print(mergedsinglekey)


df1multi = pd.DataFrame({'key1': ['A', 'B', 'C'], 'key2': [1, 2, 3], 'value1': [10, 20, 30]})
df2multi = pd.DataFrame({'key1': ['A', 'B', 'D'], 'key2': [1, 2, 4], 'value2': [40, 50, 60]})
mergedmultikeys = pd.merge(df1multi, df2multi, on=['key1', 'key2'])
print(mergedmultikeys)


outjoin = pd.merge(df1, df2, on='key', how='outer')
print(outjoin)


injoin = pd.merge(df1, df2, on='key', how='inner')
print(injoin)


leftjoin = pd.merge(df1, df2, on='key', how='left')
print(leftjoin)


rightjoin = pd.merge(df1, df2, on='key', how='right')
print(rightjoin)


concatenrows = pd.concat([df1, df2], axis=0)
print(concatenrows)


concatecolumns = pd.concat([df1, df2], axis=1)
print(concatecolumns)


dflist = [df1, df2]
concatenatedlist = pd.concat(dflist)
print(concatenatedlist)


dfwide = pd.DataFrame({'A': ['Hello', 'bY', 'Hi'], 'B': [1, 2, 3], 'C': [4, 5, 6]})
meltdf = pd.melt(dfwide, id_vars=['A'], value_vars=['B', 'C'], var_name='Variable', value_name='Value')
print(meltdf)


dfpivot = pd.DataFrame({'A': ['Hello', 'bY', 'Hello', 'bY'], 'B': ['one', 'one', 'two', 'two'], 'C': [1, 2, 3, 4]})
pivottable = pd.pivot_table(dfpivot, values='C', index='A', columns='B', aggfunc='sum')
print(pivottable)


dfgroup = pd.DataFrame({'A': ['Hello', 'bY', 'Hello', 'bY'], 'B': ['one', 'one', 'two', 'two'], 'C': [1, 2, 3, 4]})
groupdf = dfgroup.groupby('A').agg({'C': 'sum'})
print(groupdf)


groupmultiagg = dfgroup.groupby('A').agg({'C': ['sum', 'mean', 'count']})
print(groupmultiagg)


def custom_agg(x):
    return x.max() - x.min()

groupedcustomagg = dfgroup.groupby('A').agg({'C': custom_agg})
print(groupedcustomagg)
```
