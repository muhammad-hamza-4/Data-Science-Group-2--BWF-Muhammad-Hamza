# Box Plots:
Create a box plot for a given numerical data set.
            
Generate multiple box plots in a single figure to compare different data sets.
            
Customize the box plots with titles, labels, and change the colors.


```python
import matplotlib.pyplot as plt
import numpy as np

data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(1, 2, 100)

plt.boxplot(data1)
plt.title('Box Plot')
plt.show()


plt.boxplot([data1, data2], tick_labels=['Data1', 'Data2'])
plt.title('Multiple Box Plots')
plt.show()


box = plt.boxplot([data1, data2], patch_artist=True, tick_labels=['Data1', 'Data2'])
colors = ['cyan', 'lightblue']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
plt.title('Customized Box Plots')
plt.show()
```
