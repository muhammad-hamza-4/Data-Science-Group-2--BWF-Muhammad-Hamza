# Histograms:
Create a histogram for a given numerical data set. Customize the number of bins.

Plot multiple histograms on the same figure to compare different data sets.

Normalize the histograms and add appropriate labels and titles.


```python
import matplotlib.pyplot as plt
import numpy as np


data1 = np.random.normal(0, 1, 1000)
data2 = np.random.normal(1, 2, 1000)


plt.hist(data1, bins=20, edgecolor='black')
plt.title('Histogram with Custom Bins')
plt.show()

plt.hist(data1, bins=20, alpha=0.7, label='Data1', edgecolor='black')
plt.hist(data2, bins=20, alpha=0.7, label='Data2', edgecolor='black')
plt.title('Multiple Histograms')
plt.show()


plt.hist(data1, bins=20, alpha=0.7, label='Data1', edgecolor='black', density=True)
plt.hist(data2, bins=20, alpha=0.7, label='Data2', edgecolor='black', density=True)
plt.title('Normalized Histograms')
plt.show()
```
