# Pie Charts:
Create a basic pie chart.

Customize slice colors and add a legend.

Explode a slice of the pie chart.

Add percentage labels to slices.

```python
import matplotlib.pyplot as plt


labels = ['W', 'M', 'L', 'D']
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels)
plt.title('Basic Pie Chart')
plt.show()


colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.pie(sizes, labels=labels, colors=colors)
plt.title('Pie Chart with Custom Colors')
plt.show()


explode = (.1, 0, .1, 0) 
plt.pie(sizes, explode=explode, colors=colors, autopct='%1.1f%%')
plt.title('Pie Chart with Exploded Slice')
plt.show()


plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Pie Chart with Percentage Labels')
plt.show()
```
