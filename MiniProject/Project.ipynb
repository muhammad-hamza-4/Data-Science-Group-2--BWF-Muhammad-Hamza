# Data Cleaning 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('universityenrollment.csv')
df.shape
df.head()
df.tail()
df = df.drop_duplicates()
df["pre_score"] = pd.to_numeric(df["pre_score"], errors='coerce')
df["pre_score"] = df["pre_score"].fillna(df["pre_score"].mean())
df["post_score"] = df["post_score"].fillna(df["post_score"].mean())

meanprescore = df['pre_score'].mean()
df['pre_score'] = df['pre_score'].replace('-', meanprescore)

meanpostscore = df['post_score'].mean()
df['pre_score'] = df['pre_score'].replace('NA', meanpostscore)
df["pre_requirement"] = df["pre_requirement"].str.replace('NA','None')
df













```

# Enrollment Counts

```python
plt.hist(df['enrollment_count'], bins=20, edgecolor='black')
plt.title('Distribution of Enrollment Counts')
plt.xlabel('Enrollment Count')
plt.ylabel('Frequency')
plt.show()


```

# Trend B/w Online and Classroom

```python
meanenrollment = df.groupby('course_type')['enrollment_count'].mean()
meanenrollment.plot(kind='bar', color=['blue', 'green'])
plt.title('Mean Enrollment by Course Type')
plt.xlabel('Course Type')
plt.ylabel('Mean Enrollment Count')
plt.show()



```

# Pre Score vs. Enrollment Count

```python
plt.scatter(df['pre_score'], df['enrollment_count'], alpha=0.5)
plt.title('Pre Score vs. Enrollment Count')
plt.xlabel('Pre Score')
plt.ylabel('Enrollment Count')
plt.show()
```

# Post Score vs. Enrollment Count

```python

plt.scatter(df['post_score'], df['enrollment_count'], alpha=0.5)
plt.title('Post Score vs. Enrollment Count')
plt.xlabel('Post Score')
plt.ylabel('Enrollment Count')
plt.show()
```
